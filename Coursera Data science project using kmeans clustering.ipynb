# 4. Plotting the graph

```python
ver1=bank_note['V1']
ver2=bank_note['V2']
```

```python
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(ver1,ver2,s=100,alpha=0.25)
```

# 5.applying kmeans algorithm on the dataset

```python
 from sklearn.cluster import KMeans
```

```python
Ve1_Ve2=np.column_stack((ver1,ver2))
kmns_output=KMeans(n_clusters=2).fit(Ve1_Ve2)
clusters=kmns_output.cluster_centers_
y_kmeans=kmns_output.predict(Ve1_Ve2)
```

```python
plt.scatter(ver1,ver2,c=y_kmeans,cmap='plasma') 
plt.scatter(clusters[:,0],clusters[:,1],s=100)
plt.xlabel("V1")
plt.ylabel("V2")
bank_note['kmeans_clustering']=kmns_output.labels_ 
bank_note['kmeans_clustering'].value_counts()
```
